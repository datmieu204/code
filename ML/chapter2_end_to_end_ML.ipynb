```python
import pandas as pd
import requests
import os

df = pd.read_csv("screening.csv")

os.makedirs("pdfs", exist_ok=True)

for i, row in df.iterrows():
    url = row["Link"]
    title = row["Title"]
    
    # Tên file PDF (loại bỏ ký tự đặc biệt)
    safe_title = "".join(c if c.isalnum() or c in " _-" else "_" for c in title)[:100]
    filename = os.path.join("pdfs", f"{i+1}_{safe_title}.pdf")
    
    try:
        print(f"Đang tải: {title}")
        r = requests.get(url, timeout=30)
        r.raise_for_status()
        
        # Ghi ra file PDF
        with open(filename, "wb") as f:
            f.write(r.content)
        print(f"Lưu thành công: {filename}")
    except Exception as e:
        print(f"Lỗi khi tải {url}: {e}")
```
